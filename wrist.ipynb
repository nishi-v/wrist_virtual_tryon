```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import requests
from PIL import Image
```

```python
url = "https://fsvirtualnode.clouddeploy.in/coordinates"

payload = {}
files=[
  ('image',('sample_img/temp_image_50947bc4-d213-4000-9f12-55bad72e2a2e.jpg',open('sample_imgs/temp_image_6dcca584-0183-45da-beaa-56b2378cd13c.jpg','rb'),'image/jpeg'))
]
headers = {}

response = requests.request("POST", url, headers=headers, data=payload, files=files)
```

```python
print(response.text)
```

```python
result = response.text
```

```python
result
```

```python
img = Image.open('sample_imgs/temp_image_6dcca584-0183-45da-beaa-56b2378cd13c.jpg')
plt.imshow(img)
```

```python
import json

data = json.loads(result)
```

```python
wrist_data = data['results']['wrist']
```

```python
wrist_data
```

```python
from typing import List

# Separating out the coordinates
left_coords: List[float] = wrist_data["left"]
right_coords: List[float] = wrist_data["right"]
center_coords: List[float] = wrist_data["center"]
rotation_angle: float = wrist_data["rotation_angle"]
polygon_coords: List[List[float]] = wrist_data["polygon"]

# Print the separated coordinates
print("Left coordinates:", left_coords)
print("Right coordinates:", right_coords)
print("Center coordinates:", center_coords)
print("Rotation Angle:", rotation_angle)
print("Polygon coordinates:", polygon_coords)
```

```python
center_coords_revised: List[float] = [(left_coords[0] + right_coords[0]) / 2, (left_coords[1] + right_coords[1]) / 2]
```

```python
center_coords_revised
```

```python
img.size
```

```python
img_height, img_width = img.size

left_pixel = (int(left_coords[0] * img_height), int(left_coords[1] * img_width))

right_pixel = (int(right_coords[0] * img_height), int(right_coords[1] * img_width))

center_pixel = (int(center_coords_revised[0] * img_height), int(center_coords_revised[1] * img_width))

angle = int(rotation_angle)
```

```python
left_pixel, right_pixel, center_pixel
```

```python
# Convert PIL image to NumPy array
img = np.array(img)

# Draw circles at the coordinates
cv2.circle(img, left_pixel, 5, (0, 0, 255), -1)  
cv2.circle(img, center_pixel, 5, (0, 0, 255), -1)
cv2.circle(img, right_pixel, 5, (0, 0, 255), -1)  

img = Image.fromarray(img)

# Show the image with plotted points
plt.imshow(img)
plt.title('Image with Coordinates')
plt.show()
```

```python
# Convert polygon coordinates to pixel coordinates
img = np.array(img)
polygon_pixels = [(int(coord[0] * img_height / 100), int(coord[1] * img_width / 100)) for coord in polygon_coords]

cv2.polylines(img, [np.array(polygon_pixels)], isClosed = True, color=(0, 255, 0), thickness = 2)

img = Image.fromarray(img)

plt.imshow(img)
```
